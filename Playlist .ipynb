## **Relation between 'artist_genres' and 'popularity'**

```python
W1=W[['artist_genres','track_popularity','artist_name','track_name']]
W1.head()
```

#### **Track popularity greater than equal to 90**

```python
Most_popular_song=W1[W1['track_popularity'].ge(90)]
Most_popular_song
```

### **This shows number of songs with respect to track popularity**

```python
plt.hist(Most_popular_song['track_popularity'], bins=10, edgecolor='black')
plt.xlabel('track_popularity')
plt.ylabel('number of songs')
plt.show()
```

### **This shows most popular genre above 'Track popularity' 90**

```python
E=Most_popular_song['artist_genres'].value_counts().to_frame('count')
E.reset_index(inplace=True)
```

```python
plt.bar(x=E['index'],height=E['count'])
plt.xlabel('genre')
plt.ylabel('number of songs')
plt.title('most popular album')
plt.xticks(rotation=90)
plt.show()
```
